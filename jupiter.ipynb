```python
import cv2
import mediapipe as mp
import os
import time
```

```python
# Camera settings
wcap, hcap = 860, 520
cap = cv2.VideoCapture(0)
cap.set(3, wcap)
cap.set(4, hcap)

# Load overlay images
folder_path = "New folder"
mylist = os.listdir(folder_path)
overlist = []
for image_path in mylist:
    image = cv2.imread(f"{folder_path}/{image_path}")
    overlist.append(image)

# Initialize MediaPipe Hands
mp_hands = mp.solutions.hands
hands = mp_hands.Hands(max_num_hands=1, min_detection_confidence=0.75)
mp_draw = mp.solutions.drawing_utils

tipIds = [4, 8, 12, 16, 20]

ptime = 0
while True:
    success, img = cap.read()
    if not success:
        continue

    # Convert the image color from BGR to RGB
    imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # Process the image and find hands
    results = hands.process(imgRGB)

    # Check if any hand is detected
    if results.multi_hand_landmarks:
        for hand_landmarks in results.multi_hand_landmarks:
            lmList = []
            for id, lm in enumerate(hand_landmarks.landmark):
                h, w, c = img.shape
                cx, cy = int(lm.x * w), int(lm.y * h)
                lmList.append([id, cx, cy])

            fingers = []
            # Thumb
            if lmList[tipIds[0]][1] > lmList[tipIds[0] - 1][1]:
                fingers.append(1)
            else:
                fingers.append(0)

            # 4 Fingers
            for id in range(1, 5):
                if lmList[tipIds[id]][2] < lmList[tipIds[id] - 2][2]:
                    fingers.append(1)
                else:
                    fingers.append(0)

            totalFingers = fingers.count(1)
            print(totalFingers)

            # Overlay image based on the count
            h, w, c = overlist[totalFingers - 1].shape
            img[0:h, 0:w] = overlist[totalFingers - 1]
            # Display count
            cv2.rectangle(img, (20, 225), (170, 425), (0, 255, 0), cv2.FILLED)
            cv2.putText(img, str(totalFingers), (45, 375), cv2.FONT_HERSHEY_PLAIN,
                        10, (255, 0, 0), 25)

    # Calculate FPS
    ctime = time.time()
    fps = 1 / (ctime - ptime)
    ptime = ctime
    cv2.putText(img, f"FPS: {int(fps)}", (400, 70), cv2.FONT_HERSHEY_PLAIN, 3, (0, 0, 255), 3)

    # Show the image
    cv2.imshow("Image", img)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
